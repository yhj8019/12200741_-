```python
from sklearn.feature_extraction.text import TfidfVectorizer
import pandas as pd

# 데이터 그룹화
grouped = df_ess.groupby('설비번호')

# TF-IDF 벡터라이저 설정

#스탑 워드는 벡터화에 포함시키지 않겠다 (이러한 단어들은 군집화의 고려대상에서 빼겠다.)
custom_stop_words = ['발생', '불량', '마모', '측', '층', '동작불량', '부식', '오동작', '안됨', '파손', '걸림', '않됨', '탈락', 
                     '동작', '동작불량', '동작안됨', '오지시', '분리됨', '동작않됨', '위치','변형','막힘', 'leak', 'crack','pinhole','pyrites','hopper','hpr','pyrite','trouble']
tfidf_vectorizer = TfidfVectorizer(stop_words=custom_stop_words)

grouped_dfs = {}
silhouette_scores_summary = []  # 실루엣 스코어 요약 정보를 저장할 리스트

for group_name, group_df in grouped:
    
    # 텍스트 데이터 추출
    text_data = group_df['교정TM제목'].values
    
    # TF-IDF 변환
    tfidf_matrix = tfidf_vectorizer.fit_transform(text_data)
    grouped_dfs[group_name] = (group_df, tfidf_matrix)
```

```python
from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score, silhouette_samples
import matplotlib.pyplot as plt
import pandas as pd

# Truncated SVD를 사용한 차원 축소
def reduce_dimensions_svd(tfidf_matrix, n_components=10):
    n_features = tfidf_matrix.shape[1]
    n_components = min(n_components, n_features)  # n_features보다 클 수 없음
    svd = TruncatedSVD(n_components=n_components, random_state=42)
    reduced_matrix = svd.fit_transform(tfidf_matrix)
    return reduced_matrix

silhouette_scores_summary = []  # 실루엣 스코어 요약 정보를 저장할 리스트

for group_name, (group_df, tfidf_matrix) in grouped_dfs.items():
    num_samples = len(group_df)
    
    if num_samples == 1:
        group_df['Cluster'] = 0  # 하나의 클러스터로 지정
    elif num_samples == 2:
        group_df['Cluster'] = [0, 1]  # 두 개의 클러스터로 지정
    elif num_samples < 4:
        optimal_clusters = num_samples
        kmeans = KMeans(init='k-means++', n_clusters=optimal_clusters, random_state=42)
        kmeans.fit(tfidf_matrix)
        group_df['Cluster'] = kmeans.labels_
    else:
        # Truncated SVD를 사용하여 차원 축소 (벡터화 데이터는 차원이 매우 커서 군집화 성능이 떨어질것이기 때문)
        reduced_matrix = reduce_dimensions_svd(tfidf_matrix, n_components=10)
        silhouette_scores = []  # 실루엣 점수를 저장할 리스트
        K = range(2, min(11, num_samples))
        
        # 클러스터를 2 ~ 12개(12개보다 데이터수가 작다면, 데이터 수만큼) 로 나눠가며, 실루엣 스코어가 가장 높은 군집 개수를 선택
        for k in K: 
            kmeans = KMeans(init='k-means++', n_clusters=k, random_state=42)
            kmeans.fit(reduced_matrix) 
            
            if k < num_samples:
                silhouette_scores.append(silhouette_score(reduced_matrix, kmeans.labels_))
        
        # 실루엣 점수 시각화
        plt.figure(figsize=(10, 5))
        plt.plot(K[:len(silhouette_scores)], silhouette_scores, 'bo-')
        plt.xlabel('Number of clusters')
        plt.ylabel('Silhouette Score')
        plt.title(f'Silhouette Analysis for {group_name}')
        plt.show()
        
        optimal_clusters = K[silhouette_scores.index(max(silhouette_scores))]

        #최적 군집개수로 클러스터링 수행
        kmeans = KMeans(init='k-means++', n_clusters=optimal_clusters, random_state=42)
        kmeans.fit(reduced_matrix)
        group_df['Cluster'] = kmeans.labels_
        
        # 데이터 개수가 5개 이상인 하위 설비 대상으로, 군집 당 실루엣 스코어 계산
        if num_samples >= 5:
            
            sample_silhouette_values = silhouette_samples(reduced_matrix, kmeans.labels_)

            for i in range(optimal_clusters):
                cluster_silhouette_values = sample_silhouette_values[kmeans.labels_ == i]
                silhouette_scores_summary.append({
                    'group_name': group_name,
                    'num_clusters': optimal_clusters,
                    'cluster': i,
                    'silhouette_score': cluster_silhouette_values.mean()
                })
        
    grouped_dfs[group_name] = (group_df, tfidf_matrix)

# 데이터가 5개 이상인 하위 설비의 실루엣 스코어 요약 정보를 데이터프레임으로 변환
if silhouette_scores_summary:
    silhouette_scores_df = pd.DataFrame(silhouette_scores_summary)

    # 실루엣 스코어 요약 정보를 엑셀 파일로 저장
    silhouette_scores_df.to_excel('실루엣스코어_요약.xlsx', index=False)

```

```python
# 실루엣 점수가 0.5 이상인 군집들만 남긴 데이터프레임을 그룹별로 생성
filtered_grouped_dfs = {}

for group_name, (group_df, tfidf_matrix) in grouped_dfs.items():
    if len(group_df) >= 5:  # 데이터가 5개 이상인 경우에만 필터링 진행
        # 해당 그룹의 실루엣 스코어 요약 정보를 가져옴
        group_silhouette_scores = silhouette_scores_df[silhouette_scores_df['group_name'] == group_name]
        
        # 실루엣 점수가 0.5 이상인 클러스터만 선택
        valid_clusters = group_silhouette_scores[group_silhouette_scores['silhouette_score'] >= 0.5]['cluster']
        
        # Print debug information
        print(f"Group: {group_name}")
        print(f"Valid clusters (before conversion): {valid_clusters}")
        print(f"Cluster column types: {group_df['Cluster'].dtype}")
        
        # Ensure valid_clusters is of type int
        valid_clusters = valid_clusters.astype(int)
        
        # 해당 클러스터에 속한 데이터만 필터링
        filtered_group_df = group_df[group_df['Cluster'].isin(valid_clusters)]
        
        # 필터링된 데이터프레임과 원본 TF-IDF 행렬을 저장
        filtered_grouped_dfs[group_name] = (filtered_group_df, tfidf_matrix)

        # 각 클러스터별 '발행일자'의 간격을 계산하여 '최적점검주기', '최소 일자', '최대 일자', '1분위수', '2분위수', '3분위수', '4분위수' 열에 추가
        filtered_group_df['발행일자'] = pd.to_datetime(filtered_group_df['발행일자'])
        optimal_check_intervals = []
        
        for cluster in filtered_group_df['Cluster'].unique():
            cluster_df = filtered_group_df[filtered_group_df['Cluster'] == cluster]
            if len(cluster_df) > 1:
                cluster_df = cluster_df.sort_values('발행일자')
                date_diffs = cluster_df['발행일자'].diff().dropna().dt.days
                mean_interval = date_diffs.mean()
                min_interval = date_diffs.min()
                max_interval = date_diffs.max()
                q1 = date_diffs.quantile(0.25)
                q2 = date_diffs.quantile(0.50)
                q3 = date_diffs.quantile(0.75)
                q4 = date_diffs.quantile(1.00)
            else:
                mean_interval = None  # 클러스터 내 데이터가 하나일 경우, 간격을 계산할 수 없음
                min_interval = None
                max_interval = None
                q1 = None
                q2 = None
                q3 = None
                q4 = None
            
            # 실루엣 스코어에 따른 최적점검주기 할당
            silhouette_score = group_silhouette_scores[group_silhouette_scores['cluster'] == cluster]['silhouette_score'].values[0]
            if 0.5 <= silhouette_score < 0.7:
                optimal_check_interval = q1
            elif 0.7 <= silhouette_score:
                optimal_check_interval = q2
            else:
                optimal_check_interval = None  # 해당 조건 외에는 None으로 설정
            
            optimal_check_intervals.append((cluster, optimal_check_interval, min_interval, max_interval, q1, q2, q3, q4))
        
        optimal_check_intervals_df = pd.DataFrame(optimal_check_intervals, columns=['Cluster', '최적점검주기', '최소 일자', '최대 일자', '1분위수', '2분위수', '3분위수', '4분위수'])
        filtered_group_df = filtered_group_df.merge(optimal_check_intervals_df, on='Cluster', how='left')
        
        # 필터링된 데이터프레임과 원본 TF-IDF 행렬을 업데이트
        filtered_grouped_dfs[group_name] = (filtered_group_df, tfidf_matrix)

# 필터링된 결과 출력 (선택 사항)
for key, (df, _) in filtered_grouped_dfs.items():
    print(f'Group: {key}')
    print(df)

# 필요한 경우 필터링된 결과를 엑셀 파일로 저장
with pd.ExcelWriter('필터링된_결과.xlsx') as writer:
    for group_name, (filtered_group_df, _) in filtered_grouped_dfs.items():
        # Convert group_name to string to avoid errors with Excel sheet names
        sheet_name = str(group_name)
        filtered_group_df.to_excel(writer, sheet_name=sheet_name, index=False)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.font_manager as fm

# 폰트 설정 (한글이 포함된 경우 적절한 한글 폰트를 지정해야 함)
plt.rc('font', family='Malgun Gothic')  # 또는 'NanumGothic' 등 시스템에 설치된 폰트 사용

# 10605580 그룹의 군집 4에 대해 '발행일자' 간격 시각화
group_name = 10605580
cluster_label = 4

# 해당 그룹과 군집의 데이터프레임 추출
group_df, tfidf_matrix = grouped_dfs[group_name]
cluster_df = group_df[group_df['Cluster'] == cluster_label]

# '발행일자'를 datetime 형식으로 변환
cluster_df['발행일자'] = pd.to_datetime(cluster_df['발행일자'])

# '발행일자' 간격 계산
cluster_df = cluster_df.sort_values('발행일자')
date_diffs = cluster_df['발행일자'].diff().dropna().dt.days

# 스타일 설정
sns.set(style="whitegrid")

# 박스 플롯 시각화
plt.figure(figsize=(14, 7))
box_plot = sns.boxplot(x=date_diffs, color="#87CEEB", fliersize=5, linewidth=2.5, whis=1.5, width=0.4)

# 모든 데이터 포인트를 표시
strip_plot = sns.stripplot(x=date_diffs, color="#FF4500", size=8, jitter=True, edgecolor="gray", linewidth=1, alpha=0.6)

# 스타일 개선
plt.title(f'IQR Visualization for Group {group_name}, Cluster {cluster_label}', fontsize=20, fontweight='bold')
plt.xlabel('Days between Fault (Day)', fontsize=16)
plt.grid(True, linestyle='--', alpha=0.7)

plt.xticks(fontsize=14)
plt.yticks([])  # y축 눈금 제거 (단일 변수 박스 플롯이므로 필요 없음)

# 박스 플롯 내 개별 요소 설정
plt.gca().patch.set_facecolor('whitesmoke')  # 플롯 배경 색상 설정
plt.gca().patch.set_alpha(0.8)  # 배경 투명도 설정

# 박스 플롯의 스타일 설정
for box in box_plot.artists:
    box.set_edgecolor('black')
    box.set_facecolor('#87CEEB')
    box.set_alpha(0.5)

plt.show()

```
